```python
import cartopy.io.img_tiles as cimgt
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import cartopy.feature as cf

import numpy as np
# Use OpenStreetMap Tiles
osm_tiles = cimgt.OSM()
def plot_background(ax):
    ax.add_feature(cf.COASTLINE.with_scale('10m'),facecolor="None",edgecolor="k", linewidth=1)
    ax.add_feature(cf.LAKES.with_scale('10m'), facecolor = 'None', edgecolor='k')

ylo = 5#p3data['lat'].min()-0.25
yhi = 19#p3data['lat'].max()+0.25
xlo = 115#p3data['lon'].min()-0.25
xhi = 127#p3data['lon'].max()+0.25

def plot_ticks(ax):
    ax.set_yticks(np.arange(ylo,yhi,1), crs = ccrs.PlateCarree())
    ax.set_xticks(np.arange(xlo,xhi,1), crs = ccrs.PlateCarree())
    ax.yaxis.set_major_formatter(LatitudeFormatter())
    ax.xaxis.set_major_formatter(LongitudeFormatter())
```

```python
fig = plt.figure(figsize=(11, 4))
gs = gridspec.GridSpec(1, 3, width_ratios=[1, 1, 1], wspace=0.25)

# Shared color scale
vmin = 0
vmax = 70

# =========================
# Panel (a) — AMS
# =========================
ax1 = fig.add_subplot(gs[0])
ax1.text(0.01, 1.01, 'a)', transform=ax1.transAxes,
         fontsize=16, va='bottom', ha='left')

sc1 = ax1.scatter(
    p3data_xy['lon'], p3data_xy['alt']/1000,
    c=p3data_xy['Org_Ave_IsoK_STP'],
    vmin=vmin, vmax=vmax, cmap=lighter_jet, s=50
)

ax1.set_ylim(0, 5)
ax1.set_title('AMS OA (µg/m³)', fontsize=14)
ax1.set_ylabel('Altitude (km)', fontsize=16)
ax1.set_xlabel('Longitude (°E)', fontsize=16)
ax1.tick_params(axis='both', labelsize=14)


# =========================
# Panel (b) — MUSICAv0 CTRL
# =========================
ax2 = fig.add_subplot(gs[1])
ax2.text(0.01, 1.01, 'b)', transform=ax2.transAxes,
         fontsize=16, va='bottom', ha='left')

sc2 = ax2.scatter(
    musicav0data_xy_ctrl['lon'], musicav0data_xy_ctrl['alt']/1000,
    c=musicav0data_xy_ctrl['pom_mod'] + musicav0data_xy_ctrl['soa_mod'],
    vmin=vmin, vmax=vmax, cmap=lighter_jet, s=50
)

ax2.set_ylim(0, 5)
ax2.set_title('MUSICAv0 Control\n OA (µg/m³)', fontsize=14)
ax2.set_xlabel('Longitude (°E)', fontsize=16)
ax2.tick_params(axis='both', labelsize=14)
ax2.set_ylabel('Altitude (km)', fontsize=16, color='white')  # hidden


# =========================
# Panel (c) — MUSICAv0 EXP01
# =========================
ax3 = fig.add_subplot(gs[2])
ax3.text(0.01, 1.01, 'c)', transform=ax3.transAxes,
         fontsize=16, va='bottom', ha='left')

sc3 = ax3.scatter(
    musicav0data_xy_exp01['lon'], musicav0data_xy_exp01['alt']/1000,
    c=musicav0data_xy_exp01['pom_mod'] + musicav0data_xy_exp01['soa_mod'],
    vmin=vmin, vmax=vmax, cmap=lighter_jet, s=50
)

ax3.set_ylim(0, 5)
ax3.set_title('MUSICAv0 Exp01\n OA (µg/m³)', fontsize=14)
ax3.set_xlabel('Longitude (°E)', fontsize=16)
ax3.tick_params(axis='both', labelsize=14)
ax3.set_ylabel('Altitude (km)', fontsize=16, color='white')  # hidden


# =========================
# Shared Colorbar
# =========================
# Use the last scatter (sc3) but all share the same vmin/vmax
cbar_ax = fig.add_axes([0.35, -0.25, 0.3, 0.06])
cbar = plt.colorbar(sc3, cax=cbar_ax, orientation='horizontal')
cbar.ax.tick_params(labelsize=14)
cbar.set_label('OA (µg/m³)', fontsize=14)


# =========================
# Save
# =========================
plt.savefig('output/img/Figure10_AMSPaper.png',
            dpi=300, bbox_inches='tight')
```
