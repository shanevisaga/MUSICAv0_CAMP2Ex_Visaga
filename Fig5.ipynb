```python
import glob 

import xarray
import pandas as pd
import numpy as np

import datetime

import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import matplotlib.colors as mcolors
import matplotlib.dates as mdates

import seaborn as sns
```

```python
file_pattern = 'curtain_tracks/p3_2019-*track*.nc'
file_list = glob.glob(file_pattern)
ds = xarray.open_mfdataset(file_list,combine='by_coords')
df = ds.to_dataframe().reset_index()
df['datetime'] = pd.to_datetime(df['datetime'])

bin_edges = np.arange(100, 8000, 75)
df['altitude_bins'] = pd.cut(df['zco'], bins=bin_edges)
df_bc = df[(df['bc_mod']>=0) & (df['bc_sp2']>=0)]
grouped_mean_sp2 = df_bc.groupby('altitude_bins')['bc_sp2'].mean()
grouped_count_sp2 = df_bc.groupby('altitude_bins')['bc_sp2'].count()
grouped_med_sp2 = df_bc.groupby('altitude_bins')['bc_sp2'].median()
grouped_std_sp2 = df_bc.groupby('altitude_bins')['bc_sp2'].std()
grouped_mean_mod = df_bc.groupby('altitude_bins')['bc_mod'].mean()
grouped_count_mod = df_bc.groupby('altitude_bins')['bc_mod'].count()
grouped_std_mod = df_bc.groupby('altitude_bins')['bc_mod'].std()
grouped_med_mod = df_bc.groupby('altitude_bins')['bc_mod'].median()

bin_centers = bin_edges[:-1] + np.diff(bin_edges) / 2

df['oa_mod'] = df['soa_mod'] + df['pom_mod']
df_oa = df[(df['oa_mod']>=0) & (df['oa_ams']>=0)]
grouped_mean_oa_ams = df_oa.groupby('altitude_bins')['oa_ams'].mean()
grouped_count_oa_ams = df_oa.groupby('altitude_bins')['oa_ams'].count()
grouped_med_oa_ams = df_oa.groupby('altitude_bins')['oa_ams'].median()
grouped_std_oa_ams = df_oa.groupby('altitude_bins')['oa_ams'].std()
grouped_mean_oa_mod = df_oa.groupby('altitude_bins')['oa_mod'].mean()
grouped_count_oa_mod = df_oa.groupby('altitude_bins')['oa_mod'].count()
grouped_std_oa_mod = df_oa.groupby('altitude_bins')['oa_mod'].std()
grouped_med_oa_mod = df_oa.groupby('altitude_bins')['oa_mod'].median()

df_so4 = df[(df['so4_mod']>=0) & (df['so4_ams']>=0)]
grouped_mean_so4_ams = df_so4.groupby('altitude_bins')['so4_ams'].mean()
grouped_count_so4_ams = df_so4.groupby('altitude_bins')['so4_ams'].count()
grouped_med_so4_ams = df_so4.groupby('altitude_bins')['so4_ams'].median()
grouped_std_so4_ams = df_so4.groupby('altitude_bins')['so4_ams'].std()
grouped_mean_so4_mod = df_so4.groupby('altitude_bins')['so4_mod'].mean()
grouped_count_so4_mod = df_so4.groupby('altitude_bins')['so4_mod'].count()
grouped_std_so4_mod = df_so4.groupby('altitude_bins')['so4_mod'].std()
grouped_med_so4_mod = df_so4.groupby('altitude_bins')['so4_mod'].median()
```

```python
file_pattern = 'curtain_tracks_Exp01/p3_2019-*track*.nc'
file_list = glob.glob(file_pattern)
ds = xarray.open_mfdataset(file_list,combine='by_coords')
df = ds.to_dataframe().reset_index()
df['datetime'] = pd.to_datetime(df['datetime'])

bin_edges = np.arange(100, 8000, 75)
df['altitude_bins'] = pd.cut(df['zco'], bins=bin_edges)
df_bc = df[(df['bc_mod']>=0) & (df['bc_sp2']>=0)]
exp01_grouped_mean_sp2 = df_bc.groupby('altitude_bins')['bc_sp2'].mean()
exp01_grouped_count_sp2 = df_bc.groupby('altitude_bins')['bc_sp2'].count()
exp01_grouped_med_sp2 = df_bc.groupby('altitude_bins')['bc_sp2'].median()
exp01_grouped_std_sp2 = df_bc.groupby('altitude_bins')['bc_sp2'].std()
exp01_grouped_mean_mod = df_bc.groupby('altitude_bins')['bc_mod'].mean()
exp01_grouped_count_mod = df_bc.groupby('altitude_bins')['bc_mod'].count()
exp01_grouped_std_mod = df_bc.groupby('altitude_bins')['bc_mod'].std()
exp01_grouped_med_mod = df_bc.groupby('altitude_bins')['bc_mod'].median()

bin_centers = bin_edges[:-1] + np.diff(bin_edges) / 2

df['oa_mod'] = df['soa_mod'] + df['pom_mod']
df_oa = df[(df['oa_mod']>=0) & (df['oa_ams']>=0)]
exp01_grouped_mean_oa_ams = df_oa.groupby('altitude_bins')['oa_ams'].mean()
exp01_grouped_count_oa_ams = df_oa.groupby('altitude_bins')['oa_ams'].count()
exp01_grouped_med_oa_ams = df_oa.groupby('altitude_bins')['oa_ams'].median()
exp01_grouped_std_oa_ams = df_oa.groupby('altitude_bins')['oa_ams'].std()
exp01_grouped_mean_oa_mod = df_oa.groupby('altitude_bins')['oa_mod'].mean()
exp01_grouped_count_oa_mod = df_oa.groupby('altitude_bins')['oa_mod'].count()
exp01_grouped_std_oa_mod = df_oa.groupby('altitude_bins')['oa_mod'].std()
exp01_grouped_med_oa_mod = df_oa.groupby('altitude_bins')['oa_mod'].median()

df_so4 = df[(df['so4_mod']>=0) & (df['so4_ams']>=0)]
exp01_grouped_mean_so4_ams = df_so4.groupby('altitude_bins')['so4_ams'].mean()
exp01_grouped_count_so4_ams = df_so4.groupby('altitude_bins')['so4_ams'].count()
exp01_grouped_med_so4_ams = df_so4.groupby('altitude_bins')['so4_ams'].median()
exp01_grouped_std_so4_ams = df_so4.groupby('altitude_bins')['so4_ams'].std()
exp01_grouped_mean_so4_mod = df_so4.groupby('altitude_bins')['so4_mod'].mean()
exp01_grouped_count_so4_mod = df_so4.groupby('altitude_bins')['so4_mod'].count()
exp01_grouped_std_so4_mod = df_so4.groupby('altitude_bins')['so4_mod'].std()
exp01_grouped_med_so4_mod = df_so4.groupby('altitude_bins')['so4_mod'].median()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(12, 7))
gs = gridspec.GridSpec(nrows=1, ncols=6, width_ratios=[3,1, 3,1, 3,1], wspace=0.)
#gs = gridspec.GridSpec(nrows=1, ncols=6,width_ratios=[3,1, 3,1, 3,1] wspace=0.2)   # smaller spacing between paired plots
xlabels = ['OA (µg/m³)', 'SO₄ (µg/m³)', 'BC (µg/m³)']
xlims = [40, 10, 0.6]
titles = ['Organic Aerosol (OA)', 'Sulfate (SO₄)', 'Black Carbon (BC)']
legend_labels = [
    ('AMS OA', 'MUSICAv0 POM+SOA'),
    ('AMS SO₄', 'MUSICAv0 SO₄'),
    ('SP2 BC', 'MUSICAv0 BC')
]

datasets = [
    (grouped_mean_oa_ams, grouped_std_oa_ams, grouped_mean_oa_mod, grouped_std_oa_mod),
    (grouped_mean_so4_ams, grouped_std_so4_ams, grouped_mean_so4_mod, grouped_std_so4_mod),
    (grouped_mean_sp2, grouped_std_sp2, grouped_mean_mod, grouped_std_mod)
]

exp01_datasets = [
    (exp01_grouped_mean_oa_ams, exp01_grouped_std_oa_ams,
     exp01_grouped_mean_oa_mod, exp01_grouped_std_oa_mod),

    (exp01_grouped_mean_so4_ams, exp01_grouped_std_so4_ams,
     exp01_grouped_mean_so4_mod, exp01_grouped_std_so4_mod),

    (exp01_grouped_mean_sp2, exp01_grouped_std_sp2,
     exp01_grouped_mean_mod, exp01_grouped_std_mod)
]

exp01_counts_mod = [
    exp01_grouped_count_oa_mod,
    exp01_grouped_count_so4_mod,
    exp01_grouped_count_mod
]


counts_obs = [grouped_count_oa_ams, grouped_count_so4_ams, grouped_count_sp2]
counts_mod = [grouped_count_oa_mod, grouped_count_so4_mod, grouped_count_mod]

for i, ((mean_obs, std_obs, mean_mod, std_mod),
        (exp01_mean_obs, exp01_std_obs, exp01_mean_mod, exp01_std_mod),
        (label_obs, label_mod), xlabel, xlim) in enumerate(
        zip(datasets, exp01_datasets, legend_labels, xlabels, xlims)):

    # --- main profile subplot (same as before) ---
    ax = fig.add_subplot(gs[0, 2*i])
    ax.plot(mean_obs.values, bin_centers/1000, color='k', label=label_obs)
    ax.fill_betweenx(bin_centers/1000,
                     mean_obs.values - std_obs.values,
                     mean_obs.values + std_obs.values,
                     color='gray', alpha=0.3)

    ax.plot(mean_mod.values, bin_centers/1000, color='#0072B2',label=f'{label_mod} (Control)')
    ax.plot(mean_mod.values - std_mod.values, bin_centers/1000,
            color='#0072B2', linestyle=':', linewidth=2)
    ax.plot(mean_mod.values + std_mod.values, bin_centers/1000,
            color='#0072B2', linestyle=':', linewidth=2)

    # --- Exp01 model line ---
    ax.plot(exp01_mean_mod.values, bin_centers/1000,
            color='#D55E00', linewidth=2,
            label=f'{label_mod} (Exp01)')
    ax.plot(exp01_mean_mod.values - exp01_std_mod.values, bin_centers/1000,
            color='#D55E00', linestyle=':', linewidth=2)
    ax.plot(exp01_mean_mod.values + exp01_std_mod.values, bin_centers/1000,
            color='#D55E00', linestyle=':', linewidth=2)


    ax.text(0.01, 1.05, f'{chr(97+i)})', transform=ax.transAxes,
            fontsize=14,  va='bottom', ha='left')
    ax.set_ylim(0.13, 3)
    ax.set_xlim(0, xlim)
    ax.set_xlabel(xlabel, fontsize=14)
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.legend(loc='upper right', fontsize=8)
    if i == 0:
        ax.set_ylabel('Altitude (km)', fontsize=14)
    if i!=0:
        ax.tick_params(axis='y', labelleft=False)

    
plt.savefig(f'output/img/Figure5_AMSPaper.png',dpi=300,bbox_inches='tight')
```
